```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('data/used_car_dataset.csv')
```

```python
# observe 10 random samples
df.sample(10)
```

```python
# encode categorical features, year
le = preprocessing.LabelEncoder()
year_encoded = le.fit_transform(df['year_of_manufacture'])
df['year_encoded'] = year_encoded
print( df['year_encoded'].describe() )
print( le.inverse_transform([13]) )
```

```python
# encode categorical features, city
city_encoded = le.fit_transform(df['city'])
df['city_encoded'] = city_encoded
print( df['city_encoded'].describe() )
print( le.inverse_transform([7]) )
```

```python
# encode categorical features, fuel_type
fuel_encoded = le.fit_transform(df['fuel_type'])
df['fuel_encoded'] = fuel_encoded
print( df['fuel_encoded'].describe() )
print( le.inverse_transform([4]) )
```

```python
# transform kms_driven to numerical
fix_kms = lambda x: int( x[:-3].replace(',', '') )
df['kms_fixed'] = list( map(fix_kms, df['kms_driven']) )
print( df['kms_fixed'].describe() )
```
